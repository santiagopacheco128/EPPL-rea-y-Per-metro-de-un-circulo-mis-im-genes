```python
# Calcular area y perimetro de un circulo
# Variables A,P,r
# Formulas A = pi * r^2 y P = 2 * pi * r
# capturar el valor de r
import math
print("Clcular area y perimetro de un circulo")
r = float(input("introduce el valor del radio (r): "))
A = math.pi * (r**2)
P = 2 * math.pi * r
print(f"El area del circulo es: {A}")
print(f"El perimetro del circulo es: {P}")
print("Santiago Pacheco NL = 43")
```
